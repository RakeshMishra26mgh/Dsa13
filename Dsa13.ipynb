```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node

    def print_list(self):
        curr = self.head
        while curr:
            print(curr.data, end=" ")
            curr = curr.next
        print()

def create_new_list(list1, list2):
    new_list = LinkedList()
    p1 = list1.head
    p2 = list2.head
    while p1 and p2:
        if p1.data > p2.data:
            new_list.append(p1.data)
        else:
            new_list.append(p2.data)
        p1 = p1.next
        p2 = p2.next
    while p1:
        new_list.append(p1.data)
        p1 = p1.next
    while p2:
        new_list.append(p2.data)
        p2 = p2.next
    return new_list

def create_list_from_input():
    lst = LinkedList()
    size = input("Enter the size of the list: ")
    try:
        size = int(size)
        if size < 0:
            raise ValueError("Size cannot be negative")
    except ValueError as e:
        print("Invalid input:", e)
        return None
    for i in range(size):
        data = input(f"Enter data for node {i+1}: ")
        try:
            data = int(data)
            lst.append(data)
        except ValueError as e:
            print("Invalid input:", e)
            return None
    return lst

list1 = create_list_from_input()
if list1 is None:
    print("Cannot create list 1")
else:
    print("List 1:")
    list1.print_list()

list2 = create_list_from_input()
if list2 is None:
    print("Cannot create list 2")
else:
    print("List 2:")
    list2.print_list()

if list1 and list2:
    new_list = create_new_list(list1, list2)
    print("New list:")
    new_list.print_list()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node

    def print_list(self):
        curr = self.head
        while curr:
            print(curr.data, end=" ")
            curr = curr.next
        print()

def remove_duplicates(lst):
    curr = lst.head
    while curr and curr.next:
        if curr.data == curr.next.data:
            curr.next = curr.next.next
        else:
            curr = curr.next

def create_sorted_list_from_input():
    lst = LinkedList()
    size = input("Enter the size of the list: ")
    try:
        size = int(size)
        if size < 0:
            raise ValueError("Size cannot be negative")
    except ValueError as e:
        print("Invalid input:", e)
        return None
    for i in range(size):
        data = input(f"Enter data for node {i+1}: ")
        try:
            data = int(data)
            if lst.head is None or data >= lst.head.data:
                lst.append(data)
            else:
                new_node = Node(data)
                prev = None
                curr = lst.head
                while curr and data > curr.data:
                    prev = curr
                    curr = curr.next
                new_node.next = curr
                if prev is None:
                    lst.head = new_node
                else:
                    prev.next = new_node

        except ValueError as e:
            print("Invalid input:", e)
            return None
    return lst

lst = create_sorted_list_from_input()
if lst is None:
    print("Cannot create list")
else:
    print("Original list:")
    lst.print_list()

if lst:
    remove_duplicates(lst)
    print("Modified list:")
    lst.print_list()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_lists(a, b):
    if a == None:
        return b
    if b == None:
        return a

    curr_a = a
    curr_b = b
    dummy = Node(0)
    curr = dummy

    while curr_a and curr_b:
        if curr_a.data <= curr_b.data:
            curr.next = curr_a
            curr_a = curr_a.next
        else:
            curr.next = curr_b
            curr_b = curr_b.next
        curr = curr.next

    if curr_a:
        curr.next = curr_a
    if curr_b:
        curr.next = curr_b

    return dummy.next

def print_list(head):
    temp = head
    while temp:
        print(temp.data, end="->")
        temp = temp.next
    print("NULL")

def create_list():
    print("Enter the number of nodes:")
    n = int(input())
    if n <= 0:
        return None

    print("Enter the node values in sorted order:")
    head = Node(int(input()))
    temp = head

    for i in range(n-1):
        temp.next = Node(int(input()))
        temp = temp.next

    return head


a = create_list()
b = create_list()

print("List a:")
print_list(a)
print("List b:")
print_list(b)

head = merge_lists(a, b)
print("Merged list:")
print_list(head)

```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def reverse_list(head):
    if head == None or head.next == None:
        return head

    curr = head
    prev = None

    while curr:
        next = curr.next
        curr.next = prev
        curr.prev = next

        prev = curr
        curr = next

    return prev

def print_list(head):
    temp = head
    while temp:
        print(temp.data, end="<->")
        temp = temp.next
    print("NULL")

def create_list():
    print("Enter the number of nodes:")
    n = int(input())
    if n <= 0:
        return None

    print("Enter the node values:")
    head = Node(int(input()))
    temp = head

    for i in range(n-1):
        temp.next = Node(int(input()))
        temp.next.prev = temp
        temp = temp.next

    return head

head = create_list()

print("Original list:")
print_list(head)

head = reverse_list(head)
print("Reversed list:")
print_list(head)

```
